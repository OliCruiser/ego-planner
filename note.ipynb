## 生成 {p,v}
1. 找到碰撞轨迹段
2. 用 A* 找一条安全路径 Γ(大致方向)
3. 让轨迹绕着轨迹应该往 Γ这个方向移动，p = Γ上的点；v = normalize(p - Qi)。
4. 构造优化目标函数：
   $$J = smoothness + collision + dynamics$$
    $$J = λs Js + λc Jc + λd Jd$$
